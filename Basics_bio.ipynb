```python
from Bio import SeqIO
```

```python

#open fasta file and parse the sequence 

filename = '/Users/Study/Desktop/ncbi_dataset/Python_lerning/genomic.fasta'

for record in SeqIO.parse(filename, 'fasta'):
    sequence = record.seq
    print(sequence)
```

```python
from Bio.Seq import Seq
```

```python
reverse_comp_seq = sequence.reverse_complement()
```

```python
from Bio.SeqRecord import SeqRecord

# Define the output filename
output_filename = "reverse_complement.fasta"

# Write the reverse complement sequence to a FASTA file with a specified ID
with open(output_filename, "w") as output_handle:
    SeqIO.write(SeqIO.SeqRecord(reverse_comp_seq, id="reverse_complement"), output_handle, "fasta")
```
